```python
#import library 
import pandas as pd
```

```python
# Find the API url for JSON and add it here! 
# We are starting out with the Urban Park Ranger Animal Condition Response - Data from May through October

url = 'https://data.cityofnewyork.us/resource/s3vf-x992.json'

# Load the JSON into a dataframe. 
df = pd.read_json(url)

# View the first five
df.head()
```

```python
# get all the columns to get a sense for what we have
df.columns
```

```python
# let's limit a few of these columns for now 

df[['species_description', 'species_status', 'animal_condition', 'final_ranger_action','location', 'borough']][0:10]
```

```python
# what about the unique species from the last quarter 
df.species_description.unique()
```

```python
#what? There was a deer in the city? What's that all about ?

df.loc[df['species_description'] == 'White Tailed Deer']
```

Now, this is getting a little bit depressing. Some poor unhealthy infant deer was found in the Orchard Beach / City Island Traffic Circle, which (after plugging that into Google Maps) turns out to be a fairly rural area. Its not like some White Tailed Deer was wandering around Times Square. So, what happened to the deer? If you see the final_ranger_action column, you see a code called "ACC". Good thing that the documentation around the NYC Open Data sets is quite robust. This dataset came with an Excel Spreadsheet explaining all of the columns and any codes that might appear inside of them.  ACC stands for "Animal Care Center". So this baby deer was taken good care of. Maybe not such a sad story after all.  

```python
# Let's look into these turtles... 
df.loc[(df['species_description'] == 'Red-eared Slider Turtle') | (df['species_description'] == 'Turtle')]
```

So, on August 2, at 2:55pm, some member of the public called in a turtle sighting at the "Turtle Pond on E. Drive and E. 79th". From the row of data, it looks like this turtle was healthy - probably minding its own business at the "Turtle Pond". The responding ranger's action was "Advised and Educated Others".
